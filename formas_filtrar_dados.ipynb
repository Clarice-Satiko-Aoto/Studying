# Diferentes formas de filtrar os dados

```python
#importar a biblioteca pandas
import pandas as pd
```

```python
#Carregando o dataset
db_tw = pd.read_csv('stockerbot-export.csv', usecols=[0,1,2,3,4,5])
db_tw.head()
```

```python
#loc
db_tw.loc[0:5]
```

```python

# indice das linhas e nome da coluna
db_tw.loc[0:10, 'symbols']
```

```python
# iloc: numero das linhas E colunas
db_tw_iloc = db_tw.iloc[0:10, 0:3]
db_tw_iloc.tail()
```

```python
db_tw.loc[(db_tw['source'] == "MarketWatch") &  (db_tw['symbols'] == "BTC")] # e: &,  ou: |
```

```python
btc = db_tw[(db_tw['symbols'] =='BTC')]
btc.sample(3)
```

```python
#Query
db_tw.query("symbols == 'M'")
```

```python
#Filter
db_tw_filter = db_tw.filter(['text', 'source', 'company_names'])
db_tw_filter.sample(3)
```

```python
#isin

db_tw[db_tw['symbols'].isin(['BTC','GS', 'AKAM', 'JD'])]
```
